```python
import pandas as pd

df = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
df.head(), df.info()
```

```python
df_2020 = df[["Country Name", "2020"]].dropna()
```

```python
import matplotlib.pyplot as plt

top10 = df_2020.sort_values("2020", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10["2020"]/1e6, color="skyblue", edgecolor="black")
plt.title("Top 10 Countries by Population (2020)")
plt.ylabel("Population (millions)")
plt.xticks(rotation=45)
plt.show()
```

```python

df_countries = df[df["Country Code"].str.len() == 3]

df_2020 = df_countries[["Country Name", "2020"]].dropna()
```

```python
top10 = df_2020.sort_values("2020", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10["2020"]/1e6, color="skyblue", edgecolor="black")
plt.title("Top 10 Countries by Population (2020)")
plt.ylabel("Population (millions)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df_2020["2020"]/1e6, bins=20, color="lightgreen", edgecolor="black")
plt.title("Population Distribution Across Countries (2020)")
plt.xlabel("Population (millions)")
plt.ylabel("Number of Countries")
plt.show()
```

```python

exclude_keywords = ["income", "world", "IBRD", "IDA", "OECD", "Euro", "dividend"]
df_countries = df[~df["Country Name"].str.contains('|'.join(exclude_keywords), case=False, na=False)]

df_2020 = df_countries[["Country Name", "2020"]].dropna()
```

```python
top10 = df_2020.sort_values("2020", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10["2020"]/1e6, color="skyblue", edgecolor="black")
plt.title("Top 10 Countries by Population (2020)")
plt.ylabel("Population (millions)")
plt.xticks(rotation=45)
plt.show()
```
