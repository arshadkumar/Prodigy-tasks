```python
# === Setup & imports ===
!pip install folium==0.14.0 folium.plugins matplotlib seaborn scikit-learn --quiet

import os, gc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,5)

# === 1. Load (use nrows to avoid memory issues) ===
file_path = "/content/drive/My Drive/US_Accidents_March23.csv"  # adjust if needed
assert os.path.exists(file_path), "CSV not found at: " + file_path

# load a sample (first 500k rows) to keep runtime manageable
df = pd.read_csv(file_path, nrows=500000)
print("Loaded shape:", df.shape)
df.head()

# === 2. Basic cleaning & feature extraction ===
# Convert datetime columns to pandas datetime
for col in ["Start_Time","End_Time"]:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Create features: hour, weekday, date
df['start_hour'] = df['Start_Time'].dt.hour
df['start_weekday'] = df['Start_Time'].dt.day_name()
df['start_date'] = df['Start_Time'].dt.date

# Some datasets include Sunrise_Sunset or Civil_Twilight; ensure strings
for col in ['Sunrise_Sunset','Civil_Twilight','Nautical_Twilight','Astronomical_Twilight']:
    if col in df.columns:
        df[col] = df[col].astype(str)

# Create coarse severity category (if Severity numeric exist)
if 'Severity' in df.columns:
    df['Severity_cat'] = pd.cut(df['Severity'], bins=[0,1.5,2.5,3.5,4.5], labels=['1','2','3','4'])
else:
    # fallback if no Severity
    df['Severity_cat'] = df.get('Severity', np.nan)

# Drop rows without coordinates (can't map those)
df = df.dropna(subset=['Start_Lat','Start_Lng'], how='any').reset_index(drop=True)
print("After dropping missing coords:", df.shape)

# Quick memory housekeeping
gc.collect()

# === 3. Basic EDA visualizations ===
OUTDIR = "outputs"
os.makedirs(OUTDIR, exist_ok=True)

# a) Severity distribution
if 'Severity_cat' in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x='Severity_cat', data=df, order=sorted(df['Severity_cat'].dropna().unique()))
    plt.title("Accident severity distribution")
    plt.savefig(os.path.join(OUTDIR,'severity_dist.png'), dpi=150)
    plt.show()

# b) Accidents by hour
plt.figure(figsize=(10,4))
sns.countplot(x='start_hour', data=df, order=range(0,24))
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.savefig(os.path.join(OUTDIR,'by_hour.png'), dpi=150)
plt.show()

# c) Accidents by weekday
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
plt.figure(figsize=(8,4))
sns.countplot(x='start_weekday', data=df, order=order)
plt.title("Accidents by Weekday")
plt.savefig(os.path.join(OUTDIR,'by_weekday.png'), dpi=150)
plt.show()

# d) Top 10 contributing factors if column exists, e.g., Weather_Condition or Description
candidate_weather_cols = [c for c in df.columns if 'weather' in c.lower() or 'desc' in c.lower() or 'condition' in c.lower()]
print("Weather/descriptive columns found:", candidate_weather_cols)
if candidate_weather_cols:
    col = candidate_weather_cols[0]
    top = df[col].fillna("Unknown").value_counts().head(10)
    plt.figure(figsize=(8,4))
    sns.barplot(y=top.index, x=top.values)
    plt.title(f"Top 10 values in {col}")
    plt.savefig(os.path.join(OUTDIR,f'top10_{col}.png'), dpi=150)
    plt.show()

# === 4. Hotspot map (sample N points: 50k) ===
map_sample = df[['Start_Lat','Start_Lng']].sample(n=min(50000, len(df)), random_state=42).values.tolist()
m = folium.Map(location=[df['Start_Lat'].median(), df['Start_Lng'].median()], zoom_start=6)
HeatMap(map_sample, radius=8, blur=10, max_zoom=10).add_to(m)
m.save(os.path.join(OUTDIR,'heatmap.html'))
print("Heatmap saved to", os.path.join(OUTDIR,'heatmap.html'))

# === 5. Quick predictive model (optional) ===
# Try to predict Severity_cat using simple features
model_cols = []
# numeric features
for c in ['start_hour']:
    if c in df.columns: model_cols.append(c)
# boolean columns (many boolean flags exist)
bool_cols = [c for c in df.columns if df[c].dtype == 'bool']
model_cols += bool_cols[:10]  # pick first 10 boolean flags to keep model light
# simple categorical encoding for Sunrise_Sunset if present
if 'Sunrise_Sunset' in df.columns:
    df['sunset_flag'] = df['Sunrise_Sunset'].map({'Night':1,'Day':0})
    model_cols.append('sunset_flag')

# Prepare dataset (drop NA target)
df_model = df.dropna(subset=['Severity_cat'])
if len(df_model) > 20000:  # sample to speed up
    df_model = df_model.sample(20000, random_state=1)

# label encode target
le = LabelEncoder()
y = le.fit_transform(df_model['Severity_cat'])
X = df_model[model_cols].fillna(0).astype(float)

print("Model features:", model_cols)
if X.shape[0] > 0 and len(le.classes_) > 1:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
    clf = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=1, n_jobs=-1)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred, target_names=le.classes_))
else:
    print("Not enough data/labels to train model.")

# === 6. Save cleaned smaller CSV and reportable outputs ===
clean_path = os.path.join(OUTDIR, 'US_Accidents_sample_cleaned.csv')
df.to_csv(clean_path, index=False)
print("Saved cleaned sample:", clean_path)

# display the outputs folder listing
!ls -lh {OUTDIR}

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Weather condition analysis ---
plt.figure(figsize=(12,6))
top_weather = df['Weather_Condition'].value_counts().nlargest(10)
sns.barplot(x=top_weather.values, y=top_weather.index, palette="Blues_r")
plt.title("Top 10 Weather Conditions During Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.show()

# --- Road condition analysis (selected columns) ---
road_features = ['Amenity','Bump','Crossing','Give_Way','Junction','Railway','Roundabout','Station','Stop','Traffic_Signal']
road_counts = df[road_features].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=road_counts.values, y=road_counts.index, palette="Reds_r")
plt.title("Accidents by Road Features")
plt.xlabel("Number of Accidents")
plt.ylabel("Road Feature")
plt.show()
```
